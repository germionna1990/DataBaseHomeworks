# HW6. Аналитика Airflow

В ноутбуке представлены SQL-запросы для анализа данных клиентов, заказов и продуктов.
Запросы далее используются в Airflow DAG.


## Используемые таблицы

- customer
- product
- orders
- order_items

## Описание решения

Данные загружаются из CSV-файлов в PostgreSQL.
Аналитические SQL-запросы выполняются автоматически с помощью Apache Airflow (Docker).

Логика аналитики предварительно проверена SQL-запросами,
после чего перенесена в Airflow DAG без изменений.
